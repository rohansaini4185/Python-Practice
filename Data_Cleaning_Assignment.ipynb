```python
import pandas as pd
import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_raw = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
                   'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],'RecentDelays': [[23, 47], [], [24, 43, 87],
                                                                                          [13], [67, 32]],
                    'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

```python
df_raw
```

```python
New = df_raw["From_To"].str.split("_", n = 1, expand = True) 
 
df_raw["Origin"]= New[0].str.title()
df_raw["Destination"]= New[1].str.title() 

temp_df=df_raw[['Origin','Destination']]
  
df_raw
```

```python
temp_df
```

```python
for i in range(0,(df_raw.shape)[0]):
    if pd.isnull(df_raw.FlightNumber.loc[i,]):
        df_raw.loc[i,'FlightNumber'] = df_raw.FlightNumber.loc[i-1,] + 10
```

```python
df_raw['FlightNumber']=df_raw['FlightNumber'].astype(int)
```

```python
df_raw.drop(['From_To'],axis=1,inplace=True)
```

```python
Delays=pd.DataFrame(df_raw['RecentDelays'].values.tolist(), columns=['delay_1','delay_2','delay_3'])
```
